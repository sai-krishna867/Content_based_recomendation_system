```python
final_df.tags=final_df.tags.apply(lambda x: x.lower())
```

```python
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download the WordNet resource (needed for lemmatization)
import nltk
nltk.download('wordnet')
nltk.download('punkt')


# Create a WordNet lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to apply lemmatization to a text
def lemmatize_text(text):
    tokens = word_tokenize(text)
    #print(tokens)
    lemmatized_tokens = [lemmatizer.lemmatize(token,"v") for token in tokens]
    #print(lemmatized_tokens)
    return ' '.join(lemmatized_tokens)
```

```python
final_df['tags'] = final_df['tags'].apply(lemmatize_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=7500, stop_words='english')

# Fit and transform the documents to obtain TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(final_df.tags)

# Convert the TF-IDF matrix to a dense array
vectors = tfidf_matrix.toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
    index= final_df[final_df['original_title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_df.iloc[i[0]].original_title)
    return
```

```python
recommend('Superman Returns')
```
